```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#USA refugee
usa = "unhcr-persons-of-concern-residence-usa-csv-1.csv"
usa_df = pd.read_csv(usa)
```

```python
#Display data frame
usa_df.head()
```

```python
usa_new = usa_df.iloc[:,[0,1,2,3,10]]
usa_new= usa_new.rename(columns={"Country / territory of asylum/residence": 'Country of residence',
                                 "Refugees (incl. refugee-like situations)": 'Refugees'
                                })
usa_new= usa_new.drop(0)
usa_new.head()
```

```python
# convert the "*" to "0" in the data set to prevent type conversion error in next step
usa_new=usa_new.replace('*', 0)
```

```python
#convert the object to int in the year and population columns

usa_new['Year'] = pd.to_numeric(usa_new['Year'])
usa_new['Refugees'] = pd.to_numeric(usa_new['Refugees'])
usa_new['Total Population'] = pd.to_numeric(usa_new['Total Population'])

usa_new.dtypes
```

```python
#keep the data between year 2000 to 2015
usa_2000_15=usa_new.loc[(usa_new["Year"]>=2000) & (usa_new["Year"]<=2015),:]
usa_2000_15.head()
```

```python
usa_2000_15.to_csv("USA_cleaned_final.csv", index=False, header=True)
```
